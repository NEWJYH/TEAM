```python
# starttime 2022, 11, 29, 6, 41, 17
# year, month, day, hour, minute, sec
TIMEOBJECT = [2022, 11, 29, 7, 37, 27]
#                0   1   2  3  4   5
TEMPTIME = None

SMAX = 59
MMAX = 59
HMAX = 23
    
def func(x):
    global SMAX
    global MMAX
    global HMAX
    global TEMPTIME
    global TIMEOBJECT
    year_month_day , hour_min_sec = x.split()
    year, month, day = year_month_day.split('-')
    hour, minute, sec = hour_min_sec.split(":")
    
    # 반환형식에 맞춘 현재 기록된 시간
    thistime = f"{year}-{month}-{day} {hour}:{minute}:{sec}"
#     print('origin time')
#     print(thistime)
    
    # 로그시간이 None이거나 TEMPTIME <ㅡ 로그 단위 
    if TEMPTIME == None:
        TEMPTIME = f"{year}-{month}-{day} {hour}:{minute}:{sec}"
    
    # 시간이 같다면
    if TEMPTIME == thistime:
        year, month, day, hour, minute, sec = TIMEOBJECT
#         print('translate time')
        if hour < 10:
            hour = f"0{hour}"
        if minute < 10:
            minute = f"0{minute}"
        if sec < 10:
            sec = f"0{sec}"
        return f"{year}-{month}-{day} " +str(hour)+':'+str(minute)+":"+str(sec)
    elif TEMPTIME != thistime:
        TEMPTIME = f"{year}-{month}-{day} {hour}:{minute}:{sec}"
    
    year, month, day, hour, minute, sec = TIMEOBJECT
    sec += 1
    # 시간 올리고 내리고 작업을 해야함 ^^.
    if sec < SMAX:
        pass
    else:
        sec = 0
        minute += 1

    if minute < MMAX:
        pass
    else:
        minute = 0
        hour += 1

    if hour < HMAX:
        pass
    else:
        hour = 0
        day += 1

    TIMEOBJECT = [year, month, day, hour, minute, sec] 
    
    if hour < 10:
            hour = f"0{hour}"
    if minute < 10:
        minute = f"0{minute}"
    if sec < 10:
        sec = f"0{sec}"
    
    result = f"{year}-{month}-{day} " +str(hour)+':'+str(minute)+":"+str(sec)
    return result
```
